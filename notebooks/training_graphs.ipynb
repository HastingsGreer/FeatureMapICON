```python
import glob
def latest(path):
    path = "/".join(path.split("/")[:-1]) + "/loss*"
    path = sorted(glob.glob(path))
    print(path[-1])
    return path[-1]
```

```python
def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w
losses = [torch.load(latest(path)) for path in paths]
```

```python
start = 0
for l, path in zip(losses, paths):
    try:
        ll = np.array(l) * 1.0
    except:
        ll = np.array([j["loss"] for j in l][:])
        plt.plot(ll[start:])
    name = path.split("/")[-2]
    plt.plot(moving_average(ll, 80)[start:], label=name)
plt.grid()
    
import numpy as np
x = np.arange(15000)

#plt.ylim(2.4, 6.2)
plt.ylabel("log P(transform is inverse consistent)")
plt.legend()
```
